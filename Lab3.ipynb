```python
# Problem1_parta
# caps are discharging fully at the end of N cycles

import numpy as np
import matplotlib.pyplot as plt

# Parameters
K = 8                            # Number of terms in the summation
f_clock = 2.4e9                  # Clock frequency in Hz
T_s = 1 / f_clock                # Sampling period in seconds
Cs = 15.925e-12                  # Cs = 15.925 pF

# Frequency range for plotting
f = np.linspace(0, 5e9, 1000)    # Frequency range up to 5 GHz
z = np.exp(1j * 2 * np.pi * f * T_s) # z = exp(sT) = exp(j*2*pi*f*T) where T represents the sampling period

# Transfer function in z-domain
H_z_ = (1 - z**(-K)) / (1 - z**(-1))
H_z = (T_s /2*Cs) * H_z_         # final transfer function

# Magnitude and phase of the transfer function
H_mag = np.abs(H_z)
H_phase = np.angle(H_z, deg=True)

# Plotting magnitude response
plt.figure(figsize=(12, 6))
plt.plot(f / 1e9, 20 * np.log10(H_mag), label="Magnitude Response")
plt.title("Magnitude Response of Transfer Function in Z-Domain")
plt.xlabel("Frequency (GHz)")
plt.ylabel("Magnitude (dB)")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

# Plotting phase response
plt.figure(figsize=(12, 6))
plt.plot(f / 1e9, H_phase, label="Phase Response", color="orange")
plt.title("Phase Response of Transfer Function in Z-Domain")
plt.xlabel("Frequency (GHz)")
plt.ylabel("Phase (degrees)")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

plt.show()
```
