```python
#Problem 4

import numpy as np
import matplotlib.pyplot as plt

# Histogram data from Problem 4
counts = np.array([43, 115, 85, 101, 122, 170, 75, 146, 125, 60, 95, 95, 115, 40, 120, 242])
ideal_count = np.mean(counts)

# Calculate DNL and INL
dnl = counts / ideal_count - 1
inl = np.cumsum(dnl)

# Plotting
codes = np.arange(16)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(codes, dnl, marker='o')
plt.title('DNL vs Code')
plt.xlabel('Code')
plt.ylabel('DNL (LSB)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(codes, inl, marker='o')
plt.title('INL vs Code')
plt.xlabel('Code')
plt.ylabel('INL (LSB)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# problem 5

import numpy as np
import matplotlib.pyplot as plt

dnl = np.array([0,-0.5,0,0.5,-1,0.5,0.5,0])
ideal_lsb = 1.0
step_sizes = (1+dnl)*ideal_lsb

analog_edges = np.cumsum(np.insert(step_sizes, 0, 0))
offset_error = 0.5 * ideal_lsb
analog_edges += offset_error

full_scale_error = 0.5 * ideal_lsb
analog_edges[-1] += full_scale_error

ideal_edges = np.arange(len(analog_edges)) * ideal_lsb
codes= np.arange(len(analog_edges))

plt.figure(figsize=(12, 5))

plt.plot(analog_edges,codes,label = 'Actual Transfer Curve')
plt.plot(ideal_edges , codes, label= 'Ideal Transfer Curve')
plt.title('Digital Code vs Analog input')
plt.legend()
```
